```python
import tkinter as tk
from tkinter import ttk
root = tk.Tk()
root.title('高鐵查詢系統')
root.geometry('500x300')

##設定下拉式選單的「選項」，以供起點與終點選擇
cities = ['taipei', 'tainan', 'taichung']
sel_city = tk.StringVar()


##起點站 標題 與 選單
start1 = tk.Label(root, text = '起點站', font = 'Arial 15')
start1.grid(column = 0, row = 0)

city_dp1 = ttk.Combobox(root, textvariable = sel_city, value = cities, font = 'Arial 15')
        ##  .Combobox 多選擇方塊(下拉式選單)
city_dp1.grid(column = 1,row = 0) # .grid 以座標方式擺放


##終點站 標題 與 選單
end1 = tk.Label(root, text = '終點站', font = 'Arial 15')
end1.grid(column = 0, row = 1)

city_dp2 = ttk.Combobox(root, textvariable = sel_city, value = cities, font = 'Arial 15')
        ##  .Combobox 多選擇方塊
city_dp2.grid(column = 1,row = 1)

###  .grid(column = 1,row = 0)  以座標方式擺放該物件的位置



root.mainloop()
```

## 台灣彩卷

```python
url = 'https://www.taiwanlottery.com.tw'
html = requests.get(url)

html.raise_for_status()
print('網頁下載ok')

soup = bs4.BeautifulSoup(html.text, 'lxml')

# 找尋要查獎號的class
# class = 'contents_box02'
datatag = soup.find_all('div', class_ = 'contents_box02')
print('串列長度', len(datatag))

# for i in range(len(datatag)):
#     print('all 02', datatag[i].prettify())

# class ='ball_tx ball_yellow'
# 大樂透
balls = datatag[2].find_all('div', class_ = 'ball_tx ball_yellow')

print('開出順序:')
# print(balls)
for i in range(0, 6):
#     print(f'num{i}', end = ': ')
    print(balls[i].text, end = ' ')

print('\n大小順序:')
for i in range(6, 12):
#     print(f'num{i}', end = ': ')
    print(balls[i].text, end = ' ')
    
red_balls = datatag[2].find_all('div', class_ = 'ball_red') # {'class': 'ball_red'} 也可以
print('\n特別號:')
for i in range(len(red_balls)):
#     print(f'num{i}', end = ': ')
    print(red_balls[i].text, end = '')
print('')
```
