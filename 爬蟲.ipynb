## 抓圖

```python
from PIL import Image
from io import BytesIO
rey = requests.get(url2)
soupy = bs4.BeautifulSoup(rey.text, 'lxml')

img_tags = soupy.find_all('img')

target_size = 700

for img in img_tags:
    # 防出錯
    if 'src' in img.attrs: # attrs1ul 標籤屬性(attributes)
        img_url = img['src']
    elif 'data-src' in img.attrs:
        img_url = img['data-src']
    else:
        continue
#     print(img_url)
    img_re = requests.get(img_url)
    rs = img_re.raise_for_status()
#     print(rs)
    
    #check size
    image = Image.open(BytesIO(img_re.content)) # 文字text 圖片content
    width, height = image.size
    
    if width >= target_size or height >= target_size:
        print(f'照片連結: {img_url}')
        print(f'照片尺寸: {width} X {height}')
        print('-'*30)
```

```python
import os
rey = requests.get(url2)
print('網頁下載中...')
rey.raise_for_status() # .status_code 功用一樣
print('下載ok')

desdir = 'label' # 目錄
if os.path.exists(desdir) == False:
    os.mkdir(desdir)
    
obj_soup = bs4.BeautifulSoup(rey.text, 'lxml')
imgtags = obj_soup.select('img')
print(len(imgtags))
print(imgtags[0])
print(imgtags[0]['src']) # 下面一行的寫法功用一樣
print(imgtags[0].get('src'))
```

```python
for i in range(len(imgtags)):
    imgurl = imgtags[i]['src']
    
    if imgurl.startswith('http'):
        finalurl = imgurl
    else:
#     imgurl = imgurl.replace('https:/', '')
        finalurl = url2 + imgurl
    print('圖片下載中', finalurl)
    try:
        pic = requests.get(finalurl)
        pic.raise_for_status()
        print('下載圖片ok', finalurl)
        # 先開檔案再存圖片
        picfile = open(os.path.join(desdir, os.path.basename(imgurl)), 'wb')
        for disk in pic.iter_content(10240):
            picfile.write(disk)
            
        picfile.close()    
            
    except Exception as err:
        print('圖片下載失敗!!!')
```
